```python
from nilearn import datasets
anatomical_image = datasets.load_mni152_template()
from nilearn import plotting
plotting.plot_anat(anatomical_image, title='example template')
```

```python
plotting.plot_stat_map(loc_tmap, bg_img=loc_anat_filename, threshold=4, title="exmaple t test")
```

```python
plotting.plot_glass_brain(loc_tmap, threshold=3, plot_abs=False, colorbar=True, title="exmaple t test glass")
```
